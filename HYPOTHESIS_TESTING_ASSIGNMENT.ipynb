# Q1  What is a null hypothesis (H₀) and why is it important in hypothesis testing

ANSWER1
 1. The null hypothesis ($H_0$) is a statement that there is no
effect, no difference, or no relationship between the parameters of
interest. It represents the "status quo" or the widely accepted belief.
For example, $H_0: \mu = 10$ (the population mean is 10).

2. IMPORTANT
* $H_0$ is crucial because hypothesis testing is a proof by
contradiction.
*  entire process revolves around testing the assumption that $H_0$ is true.

*  We collect sample data and use statistical methods to determine if there is enough evidence to reject $H_0$ in favor of an alternative hypothesis ($H_a$ or $H_1$).


# Question 2: What does the significance level ($\alpha$) represent in  hypothesis testing?

answer 2

1. The significance level ($\alpha$) is the maximum probability of
committing a Type I error that a researcher is willing to accept.

2. It is the probability of rejecting the null hypothesis ($H_0$)
when it is actually true

3. Common values for $\alpha$ are 0.05 (5%), 0.01 (1%), or 0.10
(10%).

4. If the p-value of the test is less than or equal to $\alpha$ ($p \le
\alpha$), we reject $H_0$.

5. $\alpha$ defines the critical region (or rejection region) in the
sampling distribution.

# Question 4: Explain the difference between a one-tailed and twotailed test. Give an example of each.

answer 4

1. One-Tailed Test (Directional)
*  Alternative Hypothesis ($H_a$): Specifies a direction (either greater than $>$ or less than $<$ the hypothesized value).
* Rejection Region: Located entirely in one tail (either the right or the left) of the sampling distribution
* Example: Testing if the average product lifespan is greater than 5 years ($H_a: \mu > 5$).

2. Two-Tailed Test (Non-Directional)
* Alternative Hypothesis ($H_a$): States that the parameter is simply not equal to the hypothesized value ($\neq$)
* Rejection Region: Split between both tails (the far right and the far left) of the sampling distribution.
* Example: Testing if the average salary is different from $$$50,000 ($H_a: \mu \neq \$50,000$).

# Question 5: A company claims that the average time to resolve a customer complaint is 10 minutes. A random sample of 9 complaints gives an average time of 12 minutes and a standard deviation of 3 minutes. At $\alpha = 0.05$, test the claim.

answer 5

1.  This is a t-test for a single mean since the population standard
deviation ($\sigma$) is unknown and the sample size is small ($n <
30$).
*  Given: $\mu_0 = 10$, $n = 9$, $\bar{x} = 12$, $s = 3$, $\alpha =
0.05$.
*  Degrees of Freedom (df): $n - 1 = 9 - 1 = 8$.

2.  State Hypotheses
*  $H_0: \mu = 10$ (The average time is 10 minutes)

*  $H_a: \mu \neq 10$ (The average time is different from 10
minutes) $\rightarrow$ Two-tailed test


 3.  Calculate the Test Statistic ($t$)
 * The formula for the t-test statistic is:

* $$t = \frac{\bar{x} - \mu_0}{s / \sqrt{n}}$$
* $$t = \frac{12 - 10}{3 / \sqrt{9}} = \frac{2}{3 / 3} = \frac{2}{1} = 2.0$$


4. Determine the Critical Value (or P-value)
*  For a two-tailed t-test with $df = 8$ and $\alpha = 0.05$, the
critical t-values are found to be $\pm 2.306$.


5.  Make a Decision


*  Critical Value Method: Since $|t_{\text{calc}}| = 2.0$ is less than $|t_{\text{crit}}| = 2.306$, the test statistic falls in the non-rejection region.


*  Decision: Do not reject $H_0$.

6.  Conclusion

* At the 0.05 significance level, there is not enough evidence to reject
the company's claim that the average time to resolve a customer
complaint is 10 minutes.


# Question 6: When should you use a Z-test instead of a t-test?

ANSWER6

1. A Z-test for a mean should be used instead of a t-test when at
least one of the following two conditions is met

*  The population standard deviation ($\sigma$) is known.
*   The t-test is used when $\sigma$ is unknown and
the sample standard deviation ($s$) is used as an
estimate.
*
 2. The sample size ($n$) is large ($n \ge 30$).

 * Due to the Central Limit Theorem, when $n$ is large, the sampling distribution of the mean approximates a normaldistribution, and the t-distribution becomes very close to the standard normal (Z) distribution, even if $\sigma$ is unknown.

# Question 7: The productivity of 6 employees was measured before and after a training program. At $\alpha = 0.05$, test if the training improved productivity.



answer 7

1. This is a Paired Samples t-test (Dependent Samples t-test) because the same employees are measured before and after the training. We analyze the difference ($d$) between the paired observations.

* Given: $n = 6$, $\alpha = 0.05$.
*  Degrees of Freedom (df): $n - 1 = 6 - 1 = 5$.

2. Calculations
*  Mean Difference ($\bar{d}$): $\bar{d} = \frac{\sum d}{n} =
\frac{18}{6} = 3$
* Standard Deviation of Differences ($s_d$)

* $$s_d = \sqrt{\frac{\sum d^2 - \frac{(\sum d)^2}{n}}{n-1}} =
\sqrt{\frac{70 - \frac{(18)^2}{6}}{5}} = \sqrt{\frac{70 -
\frac{324}{6}}{5}}$$

* $$s_d = \sqrt{\frac{70 - 54}{5}} = \sqrt{\frac{16}{5}} = \sqrt{3.2}
\approx 1.7889$$


3. State Hypotheses

* We are testing if the training improved productivity, meaning "After"
should be greater than "Before", or $d = X_A - X_B$ should be
greater than 0.

*  $H_0: \mu_d \le 0$ (Training did not improve productivity)

* $H_a: \mu_d > 0$ (Training improved productivity)
$\rightarrow$ Right-tailed test

 4. Calculate the Test Statistic ($t$)

 * The formula for the paired t-test statistic is:
 * $$t = \frac{\bar{d} - \mu_{d0}}{s_d / \sqrt{n}}$$
(Assuming $\mu_{d0} = 0$)

$$t = \frac{3 - 0}{1.7889 / \sqrt{6}} = \frac{3}{1.7889 / 2.4495} =
\frac{3}{0.7303} \approx 4.108$$


5. Determine the Critical Value

* For a right-tailed t-test with $df = 5$ and $\alpha = 0.05$, the
critical t-value is $t_{\text{crit}} = 2.015$.

6.  Make a Decision
* Critical Value Method: Since $t_{\text{calc}} = 4.108$ is greater
than $t_{\text{crit}} = 2.015$, the test statistic falls in the
rejection region.

*  Decision: Reject $H_0$.

7.  At the 0.05 significance level, there is sufficient evidence to conclude
that the training program improved employee productivity.

# Question 8: A company wants to test if product preference is independent of gender.

answer 8
1. this is a Chi-Square Test for Independence ($\chi^2$).

2. State Hypotheses
*  $H_0$: Product preference is independent of gender.

*  $H_a$: Product preference is dependent on gender

3. Calculate Expected Frequencies ($E$)

* The expected frequency for each cell is $E = (\text{Row Total} \times \text{Column Total}) / \text{Grand Total}$.
* $E_{\text{Male, A}} = (50 \times 40) / 100 = 20$
* $E_{\text{Male, B}} = (50 \times 60) / 100 = 30$
* $E_{\text{Female, A}} = (50 \times 40) / 100 = 20$
* $E_{\text{Female, B}} = (50 \times 60) / 100 = 30$

4. Calculate the Test Statistic ($\chi^2$)

* The formula is:
$$\chi^2 = \sum \frac{(O - E)^2}{E}$$
Where $O$ is the Observed frequency and $E$ is the Expected
frequency.

5. $$\chi^2_{\text{calc}} = 5.0 + 3.333 + 5.0 + 3.333 = 16.666$$

4. Determine the Critical Value

* Degrees of Freedom (df): $df = (r - 1)(c - 1) = (2 - 1)(2 - 1) = 1$.

* For $\chi^2$ test with $df = 1$ and $\alpha = 0.05$, the critical value is $\chi^2_{\text{crit}} = 3.841$.

 7. Make a Decision

 * Critical Value Method: Since $\chi^2_{\text{calc}} = 16.666$ is greater than $\chi^2_{\text{crit}} = 3.841$, the test statistic falls in the rejection region.

*  Decision: Reject $H_0$.
