# Q1. Define Covariance and explain how it differs from Correlation in terms of scale and interpretation.

answer
1. Covariance is a statistical measure that describes the directional relationship between two variables ($X$ and $Y$). A positive covariance indicates that $X$ and $Y$ tend to increase or decrease together, whilea negative covariance indicates that as one increases, the other tends to decrease.
 * Formula: $Cov(X, Y) = \frac{\sum (X_i - \bar{X})(Y_i - \bar{Y})}{n-1}$ (for
a sample)


1.  feature of Covariance
* scale = Unbounded (Can range from $-\infty$ to $+\infty$)
* intrepation = Difficult to interpret the strength of the relationship, as the magnitude depends on the original units of measurement
*  units = Has units (e.g., if $X$ is in meters and $Y$ is in kilograms, covariance is in meter-kilograms).

2. correlation
* units = Bounded (Ranges strictly from $-1$ to $+1$)
* intrepation = Easy to interpret the strength and direction. A value closer to $\pm 1$ indicates a stronger relationship.
* units = Unitless (A standardized measure)




# Q2. What does a positive, negative, and zero covariance indicate about the relationship between two variables?

answer
1. Positive Covariance ($Cov > 0$): Indicates a positive linear relationship.
When the values of one variable ($X$) are above their mean, the values  
of the other variable ($Y$) tend to be above their mean as well. The variables move in the same direction.

2. Negative Covariance ($Cov < 0$): Indicates a negative (inverse) linear
relationship. When the values of $X$ are above their mean, the values of $Y$ tend to be below their mean. The variables move in opposite
directions.


3. Zero Covariance ($Cov \approx 0$): Indicates that there is no linear
relationship between the variables. The values of $X$ and $Y$ are independent of each other in a linear sense.



# Q3. Discuss the limitations of covariance as a measure of relationship between two variables. Why is correlation preferred in many cases?

answer
1. Limitations of covariance .

* Scale Dependence: The magnitude of covariance is directly affected by the units of the variables. If you change the units (e.g., from meters tocentimeters), the covariance value will change dramatically, even though
the relationship between the variables remains the same.

* Lack of Standardization: Because it is not bounded, it's impossible to tell
if a covariance of, say, 100 indicates a weak or strong relationship
without knowing the variables' standard deviations

2. Why Correlation is Preferred
* Correlation (specifically the Pearson correlation coefficient, $\rho$ or $r$) is preferred because it standardizes the measure of the relationship

* Standardization: Correlation is calculated by dividing the covariance by
the product of the standard deviations of the two variables, effectively
removing the units.

* $$r = \frac{Cov(X, Y)}{s_X s_Y}$$

* Ease of Interpretation: The resulting value is always between $-1$ and
$+1$, allowing for immediate and universal interpretation of the  strength and direction of the linear relationship, regardless of the variables' original units.


# Q4. Explain the difference between Pearson's correlation coefficient and Spearman's rank correlation coefficient. When would you prefer to use Spearman's correlation?

answer

1. feature of Pearson's Correlation Coefficient (r)

* measure = Linear relationship between the actual data values.

* Assumption =  data is approximately normally distributed and the relationship is linear (Parametric test).

* data type = interval or Ratio scale data

2. feature of Spearman's Rank Correlation Coefficient (Ï or rs)

* measure = Monotonic relationship (consistency of direction) between the ranks of the data.

* assumption = Makes no assumptions about the distribution of the data (Non
parametric test).

* data type = Ordinal, Interval, or Ratio scale data.


3. When to Prefer Spearman's Correlation:

* The relationship is non-linear but monotonic: The variables consistently
move in the same direction, but not at a constant rate (e.g., $Y$ increases slowly, then quickly, as $X$ increases).

*  The data is ordinal: The data consists of ranks (e.g., quality ratings, finish order in a race)

* The data contains outliers: Pearson's $r$ is highly sensitive to outliers,
while Spearman's $r_s$ is more robust since it uses ranks, which reduces
the impact of extreme values.



# Q5. If the correlation coefficient between two variables $X$ and $Y$ is $0.85$, interpret this value in context. Can you infer causation from this value? Why or why not?

answer

1. Interpretation of $r = 0.85$

A correlation coefficient of $r = 0.85$ indicates a strong, positive linear
relationship between variables $X$ and $Y$.
* Positive: As the values of $X$ increase, the values of $Y$ also tend to
increase.

*  Strong: The data points cluster closely around a straight line, suggesting
that the linear model is a good fit for the relationship.

2. Can you infer causation?

* No, you cannot infer causation from a correlation value alone.

*  Correlation only measures the degree to which two variables are linearly associated (they move together). It does not tell us why they
are associated.

* The relationship could be due to a third, unobserved variable (a confounding variable) that influences both $X$ and $Y$.

* This concept is summarized by the phrase: "Correlation does not imply
causation.
