#6  Explain why the uniform distribution is a good model for the outcome of rolling a fair die

ANSWER 6

1. A uniform distribution means that all outcomes are equally likely and  there is no bias toward any particular number — the die is balanced, so the chance of rolling a 1 is the same as rolling a 6.


2.  Discrete uniform distribution: Used when outcomes are finite and equally likely like die rolls.
*  Possible values: {1, 2, 3, 4, 5, 6}
*  Probability for each is 1/6



3.  Continuous uniform distribution Used when values can take any number in an interval like Rand in Excel  which generates values between 0 and 1. For dice, we use the discrete uniform distribution.




# Q8 How does log transformation help in stabilizing variance and making data more normally distributed?

ANSWER 8

1. The log transformation applies the natural logarithm (or log base 10) to each data point

2. HOW IT HELPS
* Stabilizing Variance = - Many datasets like income, population sizes, or reaction times show heteroscedasticity — variance increases with the mean.

*  log transformation reduces this effect by shrinking large values, making the spread of data more consistent across the range.

3. Making Data More Normally Distributed
* - The transformed data often looks closer to a bell-shaped curve (normal distribution).

* - This improves the validity of statistical tests (t-tests, ANOVA, regression) that assume normality.
